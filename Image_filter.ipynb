```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
from scipy import signal, ndimage, misc
from scipy.ndimage import gaussian_filter
```

```python
img = cv2.imread('barbara.jpg',flags = 0)   # flags=1 for gray level images

plt.figure(figsize=(5,5));plt.axis('off');plt.title('original image')
plt.imshow(img,cmap = 'gray')
plt.show()
```

## Gradient filter

```python
# set the kernels for contour detections (gradient operator)
kernely = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernelx = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
edges_x = cv2.filter2D(img,cv2.CV_8U,kernelx)   # CV_8U stands for 1-byte unsigned integer
edges_y = cv2.filter2D(img,cv2.CV_8U,kernely)


plt.figure(figsize=(15,5))
plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(edges_x,cmap='gray')
plt.title('Gradients X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(edges_y,cmap='gray')
plt.title('Gradients Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

## Sobel filter
Sobel operators is a joint Gausssian smoothing plus differentiation operation, and it is more resistant to noise. 

```python
sobelx = cv2.Sobel(img,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=5) 
sobely = cv2.Sobel(img,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=5)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

## Averaging filter

```python
average_kernel = np.ones((10, 10))/100
average = signal.convolve2d(img, average_kernel, boundary='symm', mode='same')

plt.figure(figsize=(10,5))
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(average,cmap = 'gray')
plt.title('Averaging filter'), plt.xticks([]), plt.yticks([])
plt.show()
```

## Gaussian filter

```python
gauss_res = gaussian_filter(img, sigma=5)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(gauss_res,cmap = 'gray')
plt.title('Gaussian filter'), plt.xticks([]), plt.yticks([])
plt.show()
```
